```python
#New Syntactic Feature

import pandas as pd
import numpy as np
import pickle
def saveObject(obj,name): 
    pickle.dump(obj,open( "repo/"+name+".pkl", "wb" ))

def loadObject(name):
    obj = pickle.load( open( "repo/"+name+".pkl", "rb" ) )
    return obj
```

```python
loc = "/home/sp/nlpfinal/features/"
f1 = "WRLevel4.txt"

# /home/sp/nlpfinal/features/BitKS3.txt
# /home/sp/nlpfinal/features/WRLevel2.txt
# /home/sp/nlpfinal/features/WRLevel3.txt
# /home/sp/nlpfinal/features/WRLevel4.txt

ffile = pd.read_csv(loc+f1)
```

```python
WR2Other = loadObject("featL3")
```

```python
DC = ffile.iloc[:,6]
S = ffile.iloc[:,2]
CT = ffile.iloc[:,7]
CP = ffile.iloc[:,8]
ffile["DC/S"] = DC/S
ffile["CT/S"] = CT/S
ffile["CP/S"] = CP/S
fname = ffile["Filename"]
```

```python
WR2Other.keys()
VP_S = WR2Other["VP_S"]
NP_S = WR2Other["NP_S"]
PP_S = WR2Other["PP_S"]
Avg_NP = WR2Other["Avg_NP"]
Avg_PP = WR2Other["Avg_NP"]
Avg_VP = WR2Other["Avg_NP"]
Sbar_S = WR2Other["Sbar_S"]
treeH_S = WR2Other["treeH_S"]
Total_S = WR2Other["Total_S"]
len(fname)
```

```python
diffWf = loadObject("wordDiffFeatures3")
```

```python
diffWf.keys()
L1 = diffWf["L1"]
L2 = diffWf["L2"]
L3 = diffWf["L3"]
L4 = diffWf["L4"]
L5 = diffWf["L5"]
L6 = diffWf["L6"]

for ij in range(len(S)):
    if S[ij] == 0:
        S[ij] = 1
```

```python
L1_S = np.zeros(len(fname))
L2_S = np.zeros(len(fname))
L3_S = np.zeros(len(fname))
L4_S = np.zeros(len(fname))
L5_S = np.zeros(len(fname))
L6_S = np.zeros(len(fname))
```

```python
VP_S1 = np.zeros(len(fname))
NP_S1 = np.zeros(len(fname))
PP_S1 = np.zeros(len(fname))
Avg_NP1 = np.zeros(len(fname))
Avg_PP1 = np.zeros(len(fname))
Avg_VP1 = np.zeros(len(fname))
Sbar_S1 = np.zeros(len(fname))
treeH_S1 = np.zeros(len(fname))
Total_S1 = np.zeros(len(fname))

for i in range(len(fname)):
    key = fname[i]
    L1_S[i] = L1[key]/S[i]
    L2_S[i] = L2[key]/S[i]
    L3_S[i] = L3[key]/S[i]
    L4_S[i] = L4[key]/S[i]
    L5_S[i] = L5[key]/S[i]
    L6_S[i] = L6[key]/S[i]
    VP_S1[i] = VP_S[key] 
    NP_S1[i] = NP_S[key]
    PP_S1[i] = PP_S[key]
    Avg_NP1[i] = Avg_NP[key]
    Avg_PP1[i] = Avg_PP[key]
    Avg_VP1[i] = Avg_VP[key]
    Sbar_S1[i] = Sbar_S[key]
    treeH_S1[i] = treeH_S[key]
```

```python
ffile["L1/S"] = L1_S
ffile["L2/S"] = L2_S
ffile["L3/S"] = L3_S
ffile["L4/S"] = L4_S
ffile["L5/S"] = L5_S
ffile["L6/S"] = L6_S
ffile["VP/S"] = VP_S1
ffile["NP/S"] = NP_S1
ffile["PP/S"] = PP_S1
ffile["Avg_NP"] = Avg_NP1
ffile["Avg_PP"] = Avg_PP1
ffile["Avg_VP"] = Avg_VP1
ffile["Sbar/S"] = Sbar_S1
ffile["treeH/S"] = treeH_S1
```

```python
ffile.to_csv("wr3SyntacticFeature.csv")
```
